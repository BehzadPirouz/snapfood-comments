```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Behzad\Documents\project\snapfood\snapfood.csv", sep="\t" ,encoding="utf-8")
df.head(10)
```

```python
df = df.drop(columns= ["Unnamed: 0"])
df.head()
```

```python
df.info()
```

```python
# df["tokens"] = df["comment"].str.split()
df["tokens"] = df["comment"].apply(lambda x: x.split())
df[["comment", "tokens"]].head()
```

```python
df["tokens_unique"] = df["tokens"].apply(lambda x: list(set(x)))
df[["comment", "tokens_unique"]].head()
```

```python
happy_comments = df[df["label"] == "HAPPY"]["tokens_unique"]
sad_comments = df[df["label"] == "SAD"]["tokens_unique"]
```

```python
from itertools import chain
happy_words = list(chain.from_iterable(happy_comments))
sad_words = list(chain.from_iterable(sad_comments))
```

```python
from collections import Counter
happy_counter = Counter(happy_words)
sad_counter = Counter(sad_words)
top50_happy = happy_counter.most_common(50)
top50_sad = sad_counter.most_common(50)
print("Top 50 Happy Words: ")
for word, count in top50_happy:
    print(f"{word}: {count}")
```

```python
print("\nTop 50 Sad Words: ")
for word, count in top50_sad:
    print(f"{word}: {count}")
```

```python
import hazm
print("Hazm imported")
```

```python
from hazm import Normalizer, word_tokenize, stopwords_list
stop_words = set(stopwords_list())
happy_filterred = [word for word in happy_words if word not in stop_words]
sad_filterred = [word for word in sad_words if word not in stop_words]
```

```python
happy_freq = Counter(happy_filterred)
sad_freq = Counter(sad_filterred)
top50_happy = happy_freq.most_common(50)
top50_sad = sad_freq.most_common(50)
print("\nTop 50 most frequent HAPPY words")
for word, count in top50_happy:
    print(f"{word}: {count}")
```

```python
print("\nTop 50 most frequent SAD words")
for word, count in top50_sad:
    print(f"{word}: {count}")
```

```python
top50_happy_words = [word for word, _ in top50_happy]
top50_sad_words = [word for word, _ in top50_sad]
print("top 50 HAPPY words")
print(top50_happy_words)
```

```python
print("top 50 SAD words")
print(top50_sad_words)
```

```python
df_words = pd.DataFrame({"Happy_Words": top50_happy_words, "Sad_Words": top50_sad_words})
print(df_words.head(10))
```
